---

<details>
<summary>Hint 1: Clustering coefficient formula</summary>

For undirected graphs:
$$C_i = \frac{2 \times \text{edges between neighbors}}{k_i \times (k_i - 1)}$$

where $k_i$ is the degree of node $i$.
